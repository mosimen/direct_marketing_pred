```python
binary_features={
'default':'yes',
'housing':'yes',
'loan':'yes',
'y':'yes'
}

ordinal_features={
'education':['unknown', 'primary','secondary','tertiary'],
'month':['jan','feb','mar','apr','may','jun','jul','aug','sep','oct','nov','dec']
}

norminal_features={
'job':'jobtype',
'marital':'ms',
'contact':'contact'
}
```

```python
def binary_encode(data, binary_col):
    for i,j in binary_col.items():
        data[i]=data[i].apply(lambda x: 1 if x==j else 0)
    return data

def ordinal_encode(data, ordinal_col):
    for i,j in ordinal_col.items():
        data[i]=data[i].apply(lambda x: j.index(x) if  str(x) != 'nan' else x)
    return data

def norminal_encode(data, norminal_col):
    for i,j in norminal_col.items():
        dummies=pd.get_dummies(data[i], prefix=j)
        data=pd.concat([data, dummies], axis=1)
        data=data.drop(i, axis=1)
    return data
```

```python
df.head()
```

```python
df=binary_encode(df, binary_features)
df=ordinal_encode(df, ordinal_features)
df=norminal_encode(df,  norminal_features)
```

```python
df.head()
```

```python
df['education']=df['education'].fillna(df['education'].median())
```

```python
df.isnull().sum()
```

```python
x=df.drop('y', axis=1)
y=df['y']
```

```python
## We can see that this is an unbalaced dataset, so we will use SMOTTENN or Smotomek techniques
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import AdaBoostClassifier,RandomForestClassifier,BaggingClassifier,GradientBoostingClassifier
from xgboost import XGBClassifier,XGBRFClassifier
from catboost import CatBoostClassifier

from sklearn.metrics import accuracy_score,classification_report,confusion_matrix

from collections import Counter
from imblearn.combine import SMOTEENN,SMOTETomek
sm=SMOTEENN()
sk=SMOTETomek()
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,random_state=50)

xtrain.shape,xtest.shape,ytrain.shape,ytest.shape
```

```python
Counter(ytrain)
```

```python
## SMOTEENN technique
```

```python
x_train, y_train=sm.fit_resample(xtrain, ytrain)
x_test, y_test=sm.fit_resample(xtest, ytest)
```

```python
Counter(y_train)
```

```python
##SMOTETomek technique
```

```python
x__train, y__train=sk.fit_resample(xtrain, ytrain)
x__test, y__test=sk.fit_resample(xtest, ytest)
```

```python
Counter(y__train)
```

```python
models={
AdaBoostClassifier():'AdaBoostClassifier',
RandomForestClassifier():'RandomForestClassifier',
XGBClassifier():'XGBClassifier',
XGBRFClassifier():'XGBRFClassifier',
BaggingClassifier():'BaggingClassifier',
GradientBoostingClassifier():'GradientBoostingClassifier'
}
```

```python
for model,name in models.items():
    model.fit(x__train,y__train)
```

```python
for model, name in models.items():
    print(name, (accuracy_score(y__test, model.predict(x__test)))*100)
    print(classification_report(y__test, model.predict(x__test)))
    print(confusion_matrix(y__test, model.predict(x__test)))
```
